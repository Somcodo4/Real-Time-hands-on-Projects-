```python
import matplotlib.pyplot as plt
```

```python
# Step 1: Input section
num_devices = int(input("How many devices do you want to include? "))
device_data = {}

for i in range(num_devices):
    name = input(f"\nEnter device #{i+1} name: ")
    power_watts = float(input(f"Enter power rating (in watts) for {name}: "))
    hours_per_day = float(input(f"How many hours/day is {name} used? "))

    # Energy in kWh/day
    daily_kwh = (power_watts * hours_per_day) / 1000
    monthly_kwh = daily_kwh * 30
    device_data[name] = {
        'daily_kwh': daily_kwh,
        'monthly_kwh': monthly_kwh
    }
```

```python
# Step 2: Display results
print("\nðŸ”‹ Estimated Monthly Energy Usage and Cost:")
rate_per_kwh = float(input("Enter your electricity rate ($ per kWh): "))
total_cost = 0

for device, usage in device_data.items():
    cost = usage['monthly_kwh'] * rate_per_kwh
    total_cost += cost
    print(f"{device}: {usage['monthly_kwh']:.2f} kWh/month | ${cost:.2f}")

print(f"\nðŸ’° Total Estimated Monthly Bill: ${total_cost:.2f}")
```

```python


# Step 3: Visualization
device_names = list(device_data.keys())
monthly_usage = [device_data[d]['monthly_kwh'] for d in device_names]

plt.figure(figsize=(10, 6))
plt.bar(device_names, monthly_usage, color='orange')
plt.title('Monthly Energy Consumption by Device')
plt.xlabel('Device')
plt.ylabel('kWh per Month')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
